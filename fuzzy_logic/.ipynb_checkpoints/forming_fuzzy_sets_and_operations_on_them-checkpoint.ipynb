## 1. Triangular and trapezoidal membership function

#### Triangular membership function

Function description:
                   **y = trimf(x, [a, b, c])**
                    
where the vector **x** is the basis set on which the MF is determined. Values **​​a** and **c** define the base of the triangle, **b** - its vertex.

$$
f(x, a, b, c) = 
\begin{cases} 
  0, & x < a, \\
  \frac{x - a}{b - a}, & a \leq x \leq b, \\
  \frac{c - x}{c - b}, & b \leq x \leq c, \\
  0, & x > c.
\end{cases}
$$




```python
import plot_script
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular_membership(x: np.ndarray, a: float, b: float, c: float) -> np.ndarray:
    """
    Calculate the triangular membership function for a given input.

    Args:
        x (np.ndarray): The input values for which to calculate membership.
        a (float): The left vertex of the triangle.
        b (float): The peak vertex of the triangle.
        c (float): The right vertex of the triangle.

    Returns:
        np.ndarray: The membership values corresponding to the input x.
    """
    # Calculate the left slope of the triangle
    left_slope = (x - a) / (b - a)
    # Calculate the right slope of the triangle
    right_slope = (c - x) / (c - b)
    # Combine the slopes and ensure values are non-negative
    return np.maximum(np.minimum(left_slope, right_slope), 0)

# Example usage
x = np.linspace(0, 10, 500)  # Generate 500 points between 0 and 10
y = triangular_membership(x, 1, 5, 9)  # Calculate membership values for the triangle
plot_script.plot_graphs(x, [{'y': y, 'label': 'Triangle(a=1, b=5, c=9)', 'color': 'SeaGreen', 'linestyle': '-'}], xlabel='x', ylabel='Membership')
```

#### Example with another parameter value.

```python
y = triangular_membership(x, 0, 7, 9.5)  
plot_script.plot_graphs(x, [{'y': y, 'label': 'Triangle(a=0, b=7, c=9.5)', 'color': 'SeaGreen', 'linestyle': '-'}], xlabel='x', ylabel='Membership')
```

#### Trapezoidal membership function

Function description: у = trapmf (х, [a, b, с, d])
where parameters **a** and **d** are the lower part of the trapezoid; **b** and **c** - the upper plateau of the trapezium

$$
f(x, a, b, c, d) = 
\begin{cases}
0, & x < a, \\
\frac{x-a}{b-a}, & a \leq x < b, \\
1, & b < x \leq c, \\
\frac{d-x}{d-c}, & c < x \leq d, \\
0, & x > d.
\end{cases}
$$

```python
def trapezoid_membership(x: np.ndarray, a: float, b: float, c: float, d: float) -> np.ndarray:
    """
    Calculate the trapezoidal membership function values for a given input.

    Args:
        x (np.ndarray): Input values for which to calculate membership.
        a (float): The left base of the trapezoid.
        b (float): The left peak of the trapezoid.
        c (float): The right peak of the trapezoid.
        d (float): The right base of the trapezoid.

    Returns:
        np.ndarray: Membership values corresponding to the input values.
    """
    return np.where(
        # If x is less than or equal to a or greater than or equal to d, membership is 0
        (x <= a) | (x >= d), 0,
        np.where(
            # If x is between a and b, calculate the slope from a to b
            (a < x) & (x <= b), (x - a) / (b - a),
            np.where(
                # If x is between b and c, membership is 1
                (b < x) & (x <= c), 1,
                # If x is between c and d, calculate the slope from d to c
                (d - x) / (d - c)
            )
        )
    )

# Example usage
x = np.linspace(0, 10, 500)
y = trapezoid_membership(x, 1, 4, 6, 9)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Trapezoid(a=1, b=4, c=6, d=9)', 'color': 'SeaGreen', 'linestyle': '-'}], xlabel='x', ylabel='Membership')
```

#### Example with another parameter value.

```python
x = np.linspace(0, 10, 500)
y = trapezoid_membership(x, 1, 2, 6, 9)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Trapezoid(a=1, b=2, c=6, d=9)', 'color': 'SeaGreen', 'linestyle': '-'}], xlabel='x', ylabel='Membership')
```

## 2. Simple and Two-Sided Gaussian Membership Function

### Simple Gaussian Membership Function

Function description:
y = gaussmf (x, [σ, z]

The symmetric Gaussian function depends on two paramete - **s** σ an **d** z

$$\mu(x) = e^{-\frac{(x-c)^2}{2\sigma^2}}$$

```python
def gaussmf(x, mean, sigma):
    return np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))


x = np.linspace(-10, 10, 1000)
y = gaussmf(x, 0, 3)
plot_script.plot_graphs(x, [{'y': y, 'label': 'gaussmf(mean=0, sigma=3)', 'color': 'SeaGreen', 'linestyle': '-'}], xlabel='x', ylabel='y')
```

#### Example with another parameter value.

```python
def gaussmf(x, mean, sigma):
    return np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

x = np.linspace(-10, 10, 1000)
y_gauss = gaussmf(x, 2.5, 3)

plot_script.plot_graphs(x, [{'y': y_gauss, 'label': 'Gauss MF', 'color': 'SeaGreen', 'linestyle': '-'}])
plot_script.add_vertical_line(2.5)
```

### Two-Sided Gaussian Membership Function

Function description: у = gauss2mf (x, [σ1, з1, σ2, c2])

```python
def gauss2mf(x, mean1, sigma1, mean2, sigma2):
    assert mean1 <= mean2

    return np.where(
        x <= mean1, gaussmf(x, mean1, sigma1),
        np.where(
            x >= mean2, gaussmf(x, mean2, sigma2),
            (gaussmf(x, mean1, sigma1) + gaussmf(x, mean2, sigma2)) / 2
        )
    )

y = gaussmf(x, 0, 3)
y1 = gauss2mf(x,0, 2, 0, 2)
y2 = gauss2mf(x,0, 4, 0, 4)

plot_script.plot_graphs(x, [{'y': y, 'label': 'Gauss MF', 'color': '#10646e', 'linestyle': '-'},
                           {'y': y1, 'label': 'Gauss 2MF', 'color': '#212e75', 'linestyle': '--'},
                           {'y': y2, 'label': 'Gauss 2MF', 'color': '#751d33', 'linestyle': '--'}])
```

## 3. Generalized bell shape membership function

Its difference from the previously considered FP consists in the addition of a third parameter, which allows for a smooth transition between fuzzy sets.

Function description:
                        y = gbellmf (x, [a, b, c]).

$$f(x, a, b, c) = \frac{1}{1 + \left|\frac{|x-c|}{a}\right|^{2b}}$$

```python
def gbellmf(x, a, b, c):
    return 1/(1 + np.abs(((x - c) ** (2 * b)) / a))


x = np.linspace(-5, 5, 1000)
y = gbellmf(x, 1, 2, 0)

plot_script.plot_graphs(x, [{'y': y, 'label': 'Generalized bell(1, 2, 0)', 'color': '#10646e', 'linestyle': '-'}])
```

#### Example with another parameter value.

```python
y = gbellmf(x, 1.5, 1, 2)

plot_script.plot_graphs(x, [{'y': y, 'label': 'Generalized bell(1.5, 1, 2)', 'color': '#10646e', 'linestyle': '-'}])
plot_script.add_vertical_line(2)
```

```python
y = gbellmf(x, 5, 4, -2)

plot_script.plot_graphs(x, [{'y': y, 'label': 'Generalized bell(5, 4, -2)', 'color': '#10646e', 'linestyle': '-'}])
plot_script.add_vertical_line(-2)
```

## 4. Sigmoidal membership functions

The main sigmoidal membership function is denoted by sigmf, and the additional ones by dsigmf and psigmf.

Description of the basic sigmoidal function
                    
y = sigmf (x, [a, c]).

$$f(x, a, c) = \frac{1}{1 + e^{-a(x-c)}}$$

```python
def sigmf(x, a, c):
    return 1 / (1 + np.exp(-a * (x-c)))

x = np.linspace(-10, 10, 1000)
y = sigmf(x, 2, 0)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Sigmoid MF', 'color': '#10646e', 'linestyle': '-'}])
```

```python
def dsigmf(x, a1, c1, A2, c2):
    sig1 = sigmf(x, a1, c1)
    sig2 = sigmf(x, A2, c2)
    return sig1 - sig2

y = dsigmf(x, 2, 1, 3, 6)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Diff of 2 Sigmoid MF', 'color': '#10646e', 'linestyle': '-'}])
```

```python
def psigmf(x, a1, c1, A2, c2):
    sig1 = sigmf(x, a1, c1)
    sig2 = sigmf(x, A2, c2)
    return sig1 * sig2


y = psigmf(x, 2, 1, 3, 6)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Prod of 2 Sigmoid MF', 'color': '#10646e', 'linestyle': '-'}])
```

## 5. Polynomial membership functions (Z-, PI- and S-functions)

Membership functions based on polynomial curves, the corresponding functions are called Z-functions (zmf), PI-functions (pimf) and S-functions (smf).

Function description:  
y = zmf (x, [a, b])
                    .
Parameters a and b determine the extreme values ​​of the curvhe curve

Function description:
                    y = pimf (x, [a, b, c, d]).
                    
Parameters a and d set the transition of the function to a zero value, and parameters b and c - to one.

Function description:
                    y = smf (x, [a, b]).

Parameters a and b determine the extreme values ​​of the curve

```python
def smf(x, a, b):
    return np.where((a <= x) & (x < (a + b) / 2),
                    2 * ((x - a) / (b - a)) ** 2,
                    np.where(((a + b) / 2 <= x) & (x <= b),
                             1 - 2 * ((x - b) / (b - a)) ** 2,
                             np.where(x <= a, 0, np.where(x >= b, 1, np.nan))))

x = np.linspace(-5, 5, 500)
s = smf(x, -2, 2)
plot_script.plot_graphs(x, [{'y': s, 'label': 'S-finction MF', 'color': '#10646e', 'linestyle': '-'}])
```

```python
def zmf(x, a, b):
    return np.where((a <= x) & (x < (a + b) / 2),
                    1 - 2 * ((x - a) / (b - a)) ** 2,
                    np.where(((a + b) / 2 <= x) & (x <= b),
                             2 * ((x - b) / (b - a)) ** 2,
                             np.where(x <= a, 1, np.where(x >= b, 0, np.nan))))

x = np.linspace(-5, 5, 500)
z = zmf(x, -2, 2)
plot_script.plot_graphs(x, [{'y': z, 'label': 'S-finction MF', 'color': '#10646e', 'linestyle': '-'}])
```

```python
def pimf(x, a, b, c, d):
    return np.where(x <= a, 0,
                    np.where((a <= x) & (x <= (a + b) / 2),
                             2 * ((x - a) / (b - a)) ** 2,
                             np.where(((a + b) / 2 < x) & (x <= b),
                                      1 - 2 * ((x - b) / (b - a)) ** 2,
                                      np.where((b < x) & (x <= c),
                                               1,
                                               np.where((c <= x) & (x < (c + d) / 2),
                                                        1 - 2 * ((x - c) / (d - c)) ** 2,
                                                        np.where(((c + d) / 2 <= x) & (x <= d),
                                                                 2 * ((x - d) / (d - c)) ** 2,
                                                                 np.where(x >= d, 0, np.nan)))))))

x = np.linspace(0, 10, 500)
y = pimf(x, 1, 3, 5, 7)

plot_script.plot_graphs(x, [{'y': y, 'label': 'Pi-finction MF', 'color': '#10646e', 'linestyle': '-'}])
```

Another way to represent the pi membership function is the product of the s- and z- function

```python
def pimf(s, z):
    return s*z

x = np.linspace(0, 9, 500)
y = pimf(s,z)
plot_script.plot_graphs(x, [{'y': y, 'label': 'Pi-finction MF', 'color': '#10646e', 'linestyle': '-'}])
```

## 6. Minimax interpretation of logical operators with variable operations in search of a minimum and a maximum.

Description of the conjunctive function: y = min ([в1; у2]).

Description of the disjunctive function: y = max ([в1; у2]).

```python
def min_mf(gauss1, gauss2):
    return np.fmin(gauss1, gauss2)

x = np.linspace(-20, 30, 1000)

g1 = gaussmf(x, 2, 5)
g2 = gaussmf(x, 9, 5)
y = min_mf(g1, g2)

plot_script.plot_graphs(x, [{'y': g1, 'label': 'gaussmf1', 'color': '#10646e', 'linestyle': '--'},
                           {'y': g2, 'label': 'gaussmf2', 'color': '#10646e', 'linestyle': '--'},
                           {'y': y, 'label': 'gaussmf2', 'color': '#152963', 'linestyle': '-'}])
```

```python
def max_mf(gauss1, gauss2):
    return np.fmax(gauss1, gauss2)

g1 = gaussmf(x, 2, 5)
g2 = gaussmf(x, 9, 5)
y = max_mf(g1, g2)

plot_script.plot_graphs(x, [{'y': g1, 'label': 'gaussmf1', 'color': '#10646e', 'linestyle': '--'},
                           {'y': g2, 'label': 'gaussmf2', 'color': '#10646e', 'linestyle': '--'},
                           {'y': y, 'label': 'gaussmf2', 'color': '#152963', 'linestyle': '-'}])
```

## 7. Probable interpretation of conjunctive and disjunctive operators

```python
def and_func(gauss1, gauss2):
    return gauss1 * gauss2

g1 = gaussmf(x, 2, 5)
g2 = gaussmf(x, 9, 5)
y = and_func(g1, g2)

plot_script.plot_graphs(x, [{'y': g1, 'label': 'gaussmf1', 'color': '#10646e', 'linestyle': '--'},
                           {'y': g2, 'label': 'gaussmf2', 'color': '#10646e', 'linestyle': '--'},
                           {'y': y, 'label': 'gaussmf2', 'color': '#152963', 'linestyle': '-'}])
```

```python
def or_func(gauss1, gauss2):
    return gauss1 + gauss2 - gauss1 * gauss2

g1 = gaussmf(x, 2, 5)
g2 = gaussmf(x, 9, 5)
y = or_func(g1, g2)

plot_script.plot_graphs(x, [{'y': g1, 'label': 'gaussmf1', 'color': '#10646e', 'linestyle': '--'},
                           {'y': g2, 'label': 'gaussmf2', 'color': '#10646e', 'linestyle': '--'},
                           {'y': y, 'label': 'gaussmf2', 'color': '#152963', 'linestyle': '-'}])
```

## 8. The addition of a fuzzy set "NOT A", Difference plot 

The addition of a fuzzy set is nothing but a mathematical representation of the verbal expression "NOT A", where A is a fuzzy set that describes some fuzzy judgment.

```python
def diff_plot(gauss):
    return 1-gauss


x = np.linspace(-20, 30, 1000)
g1 = gaussmf(x, 4, 10)
y = diff_plot(g1)

plot_script.plot_graphs(x, [{'y': g1, 'label': 'gaussmf1', 'color': '#10646e', 'linestyle': '--'},
                           {'y': y, 'label': 'diff gaussmf', 'color': '#152963', 'linestyle': '-'}])
```
